```python
# Create a comprehensive data cleaning function
def clean_icu_data(df):
    """
    Comprehensive data cleaning for ICU prediction dataset
    """
    df_clean = df.copy()
    
    print("=== DATA CLEANING REPORT ===\n")
    print(f"Original dataset shape: {df_clean.shape}")
    
    # 1. Handle missing values
    print("\n1. HANDLING MISSING VALUES:")
    
    # Identify columns with excessive missing values (>70%)
    missing_threshold = 0.7
    missing_pct = df_clean.isnull().sum() / len(df_clean)
    high_missing_cols = missing_pct[missing_pct > missing_threshold].index.tolist()
    
    print(f"Columns with >{missing_threshold*100}% missing values ({len(high_missing_cols)} columns):")
    for col in high_missing_cols[:10]:  # Show first 10
        print(f"  - {col}: {missing_pct[col]*100:.1f}% missing")
    if len(high_missing_cols) > 10:
        print(f"  ... and {len(high_missing_cols)-10} more columns")
    
    # Drop columns with excessive missing values
    df_clean = df_clean.drop(columns=high_missing_cols)
    print(f"Dropped {len(high_missing_cols)} columns with excessive missing values")
    
    # 2. Handle remaining missing values
    remaining_missing = df_clean.isnull().sum()
    cols_with_missing = remaining_missing[remaining_missing > 0].index.tolist()
    
    if len(cols_with_missing) > 0:
        print(f"\nRemaining columns with missing values: {len(cols_with_missing)}")
        
        # For numerical columns, fill with median
        numerical_cols = df_clean.select_dtypes(include=[np.number]).columns
        for col in cols_with_missing:
            if col in numerical_cols:
                median_val = df_clean[col].median()
                df_clean[col].fillna(median_val, inplace=True)
                print(f"  - Filled {col} missing values with median: {median_val:.3f}")
        
        # For categorical columns, fill with mode
        categorical_cols = df_clean.select_dtypes(include=['object']).columns
        for col in cols_with_missing:
            if col in categorical_cols:
                mode_val = df_clean[col].mode()[0] if not df_clean[col].mode().empty else 'Unknown'
                df_clean[col].fillna(mode_val, inplace=True)
                print(f"  - Filled {col} missing values with mode: {mode_val}")
    
    # 3. Handle data types
    print("\n2. STANDARDIZING DATA TYPES:")
    
    # Convert AGE_PERCENTIL to ordinal encoding
    age_mapping = {
        '10th': 1, '20th': 2, '30th': 3, '40th': 4, '50th': 5,
        '60th': 6, '70th': 7, '80th': 8, '90th': 9, 'Above 90th': 10
    }
    df_clean['AGE_PERCENTIL_ENCODED'] = df_clean['AGE_PERCENTIL'].map(age_mapping)
    print(f"Encoded AGE_PERCENTIL to numerical values (1-10)")
    
    # Convert WINDOW to ordinal encoding
    window_mapping = {'0-2': 1, '2-4': 2, '4-6': 3, '6-12': 4, 'ABOVE_12': 5}
    df_clean['WINDOW_ENCODED'] = df_clean['WINDOW'].map(window_mapping)
    print(f"Encoded WINDOW to numerical values (1-5)")
    
    # 4. Remove duplicates (if any)
    initial_rows = len(df_clean)
    df_clean = df_clean.drop_duplicates()
    removed_duplicates = initial_rows - len(df_clean)
    if removed_duplicates > 0:
        print(f"\n3. REMOVED DUPLICATES: {removed_duplicates} duplicate rows")
    else:
        print(f"\n3. DUPLICATES: No duplicate rows found")
    
    # 5. Data validation
    print("\n4. DATA VALIDATION:")
    
    # Check for impossible values in binary columns
    binary_cols = [col for col in df_clean.columns if df_clean[col].nunique() == 2 and set(df_clean[col].unique()).issubset({0, 1, 0.0, 1.0})]
    print(f"Identified {len(binary_cols)} binary columns")
    
    # Ensure binary columns only have 0/1 values
    for col in binary_cols:
        unique_vals = df_clean[col].unique()
        if not set(unique_vals).issubset({0, 1, 0.0, 1.0}):
            print(f"  Warning: {col} has unexpected values: {unique_vals}")
    
    # 6. Feature engineering (optional improvements)
    print("\n5. FEATURE ENGINEERING:")
    
    # Create age group from AGE_ABOVE65
    df_clean['AGE_GROUP'] = df_clean['AGE_ABOVE65'].map({0: 'Under_65', 1: 'Above_65'})
    print("Created AGE_GROUP categorical feature")
    
    # Create gender labels
    df_clean['GENDER_LABEL'] = df_clean['GENDER'].map({0: 'Female', 1: 'Male'})
    print("Created GENDER_LABEL categorical feature")
    
    print(f"\nFinal dataset shape: {df_clean.shape}")
    print(f"Columns removed: {df.shape[1] - df_clean.shape[1]}")
    print(f"Rows removed: {df.shape[0] - df_clean.shape[0]}")
    
    return df_clean

# Apply cleaning
df_cleaned = clean_icu_data(df)
```

```python
# Preprocessing pipeline
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder  # Added import for LabelEncoder
from sklearn.impute import SimpleImputer  # Added import for SimpleImputer
from sklearn.model_selection import train_test_split  # Added import for train_test_split

# Create a copy of the data for preprocessing
X_processed = X.copy()

# Handle categorical variables
le_age = LabelEncoder()
le_window = LabelEncoder()

X_processed['AGE_PERCENTIL'] = le_age.fit_transform(X_processed['AGE_PERCENTIL'])
X_processed['WINDOW'] = le_window.fit_transform(X_processed['WINDOW'])

print("Label encodings:")
print("AGE_PERCENTIL:", dict(zip(le_age.classes_, le_age.transform(le_age.classes_))))
print("WINDOW:", dict(zip(le_window.classes_, le_window.transform(le_window.classes_))))

# Handle missing values in numerical columns
imputer = SimpleImputer(strategy='median')
X_processed[numerical_cols] = imputer.fit_transform(X_processed[numerical_cols])

print(f"\nMissing values after imputation: {X_processed.isnull().sum().sum()}")

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, 
    test_size=0.2, 
    random_state=42, 
    stratify=y  # Maintain the same proportion of target classes
)

print(f"\nTraining set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
print(f"Training target shape: {y_train.shape}")
print(f"Testing target shape: {y_test.shape}")

# Check class distribution in train and test sets
print(f"\nTraining set target distribution:")
print(y_train.value_counts(normalize=True))
print(f"\nTesting set target distribution:")
print(y_test.value_counts(normalize=True))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import warnings
warnings.filterwarnings('ignore')

# Load the training and testing data
X_train = pd.read_csv('X_train_scaled.csv')
X_test = pd.read_csv('X_test_scaled.csv')
y_train = pd.read_csv('y_train.csv').iloc[:, 0]  # Get the first column
y_test = pd.read_csv('y_test.csv').iloc[:, 0]    # Get the first column

print("Data loaded successfully!")
print(f"Training set: {X_train.shape}, Target: {y_train.shape}")
print(f"Testing set: {X_test.shape}, Target: {y_test.shape}")

# Train Random Forest Classifier
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    class_weight='balanced'  # Handle class imbalance
)

print("\nTraining Random Forest model...")
rf_model.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_model.predict(X_test)
y_pred_proba_rf = rf_model.predict_proba(X_test)[:, 1]

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred_rf)
precision = precision_score(y_test, y_pred_rf)
recall = recall_score(y_test, y_pred_rf)
f1 = f1_score(y_test, y_pred_rf)

print(f"\n🎯 MODEL PERFORMANCE METRICS:")
print(f"   • Accuracy: {accuracy:.3f}")
print(f"   • Precision: {precision:.3f}")
print(f"   • Recall: {recall:.3f}")
print(f"   • F1-Score: {f1:.3f}")
```
